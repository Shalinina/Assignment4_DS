```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None

class Stack:
    def __init__(self):
        self.top = None
        self.size = 0

    def push(self, value):
        new_node = Node(value)
        new_node.next = self.top
        self.top = new_node
        self.size += 1

    def pop(self):
        if self.is_empty():
            return None
        value = self.top.value
        self.top = self.top.next
        self.size -= 1
        return value

    def is_empty(self):
        return self.size == 0

    def __repr__(self):
        if self.is_empty():
            return "<Stack (0 elements): []>"
        if self.size == 1:
            return f"<Stack (1 element): [{self.top.value}]>"
        elif self.size == 2:
            return f"<Stack (2 elements): [{self.top.value}, {self.top.next.value}]>"
        elif self.size == 3:
            return f"<Stack (3 elements): [{self.top.value}, {self.top.next.value}, {self.top.next.next.value}]>"
        else:
            current = self.top
            elements = []
            while current:
                elements.append(str(current.value))
                current = current.next
            return f"<Stack ({self.size} elements): {elements}>"

def check_balance(text):
    stack = Stack()
    opening_brackets = "({["
    closing_brackets = ")}]"
    pairs = 0

    for i, char in enumerate(text):
        if char in opening_brackets:
            stack.push((char, i))
        elif char in closing_brackets:
            if stack.is_empty():
                return f"Match error at position {i}"
            else:
                top = stack.pop()
                if (top[0] == '(' and char != ')') or \
                   (top[0] == '[' and char != ']') or \
                   (top[0] == '{' and char != '}'):
                    return f"Match error at position {i}"
                pairs += 1

    if not stack.is_empty():
        return f"Match error at position {stack.pop()[1]}"

    return f"Ok - {pairs}"

# Test the code
#print(check_balance("a(b)c[d]e{f}g"))
# print(check_balance("a(b)c[)d]e{f}g"))
# print(check_balance("a(b)(((c[d]e{f}g)))"))
# print(check_balance("a(b)c(([d][e{f}])g)"))
# print(check_balance("a(b)c(([d][e{f}])g)("))
# print(check_balance("]a(b)c(([d][e{f}])g)("))
# print(check_balance("abc"))
```
